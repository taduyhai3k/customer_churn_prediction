```python
import mydataset as dp
import LR
import warnings
import torch
from sklearn.exceptions import UndefinedMetricWarning

# Bỏ qua cảnh báo
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)
```

```python
train = dp.pd.read_csv("churn-bigml-80.csv")
test = dp.pd.read_csv("churn-bigml-20.csv")
```

```python
train = dp.MCI_Dataset("D:\Học tập\Học kì 20231\Linh tinh\MCI Test", "churn-bigml-80.csv")
test = dp.MCI_Dataset("D:\Học tập\Học kì 20231\Linh tinh\MCI Test", "churn-bigml-20.csv")
```

```python
train_dataloader = dp.DataLoader(train, batch_size=32, shuffle=True)
test_dataloader = dp.DataLoader(test, batch_size=32, shuffle = True)
```

```python
LR_simple = LR.LogisticRegression(19)
```

```python
optimizer = LR.optim.Adam(LR_simple.parameters(), lr = 0.001)
for i in range(50):
    print(f"epoch {i}")
    train_loss = LR.train_loop(train_dataloader, LR_simple, optimizer)
    test_loss = LR.test_loop(test_dataloader, LR_simple)
    print(train_loss, test_loss)
```
