```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv("churn-bigml-80.csv")
df1_raw = pd.read_csv("churn-bigml-20.csv")
```

```python
df_raw.describe()
```

```python
df_raw.head(5)
```

```python
encode_columns = [col for col in df_raw.columns if df_raw[col].dtype == 'O'or df_raw[col].dtype == 'bool' ]    
label_encoder = LabelEncoder()
for col in encode_columns:
    df_raw[col] = label_encoder.fit_transform(df_raw[col])
    df1_raw[col] = label_encoder.fit_transform(df1_raw[col])
```

```python
df_raw.head(5)
```

```python
fig, axs = plt.subplots(4,5, figsize = (30, 16))
j = 0
k = 0
for i in df_raw.columns:
    sns.histplot(data=df_raw[i], ax=axs[j,k])
    k = k+1
    k = k % 5
    if k == 0:
        j = j+1
```

```python
correlation_matrix = df_raw.corr()
plt.figure(figsize=(30, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".1f")
plt.title("Correlation Matrix")
plt.show()
```

```python
drop_columns = ["State", "Area code", "Number vmail messages", "Total day minutes", "Total eve minutes", "Total night minutes", "Total intl minutes"]
```

```python
df_new = df_raw.drop(columns=drop_columns)
```

```python
df1_new = df1_raw.drop(columns=drop_columns)
```

```python
df_new.to_csv("train.csv", index = False)
df1_new.to_csv("test.csv", index = False)
```
